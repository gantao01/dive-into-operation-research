```python
def Preprocess(file):
    array = []
    for line in f:
        array.append(line)
    xc = [0]*300
    yc = [0]*300
    maxrows = 300
    maxcols = 300
    Cost = [([0] * maxcols) for row in range(maxrows) ]

    dim = int(array[0])
    for it in range(1, dim+1):
        p = array[it]
        tokens = p.split()
        xc[it-1] = float(tokens[0])
        yc[it-1] = float(tokens[1])
    
    for i in range(0, dim):
        for j in range(0, dim):
            Cost[i][j] = pow((pow(xc[i]-xc[j], 2) + pow(yc[i] - yc[j], 2)), 0.5)
        
    for i in range(0, dim):
        for j in range(0, dim):
            if(i==j):
                Cost[i][j] = 99999
    return Cost, xc, yc, dim
```

```python
from gurobipy import *
import string
import time

f = open("C:\\Users\\Prasangsha\\Downloads\\TSP_instance_n_30_s_1.dat")
Cost, xc, yc, dim = Preprocess(f)

startTime = time.time()
TSP_SP = Model('TSP_Shortest_Path')


x = [[[0 for i in range(dim)] for j in range(dim)] for t in range(dim)] 
for i in range(0, dim):
    for j in range(0, dim):
        for t in range(0, dim):
            x[i][j][t] = TSP_SP.addVar(vtype=GRB.BINARY, name = 'x[][][]'.format(i,j,t))
    
TSP_SP.update()

TSP_SP.setObjective(quicksum(Cost[i][j]*(quicksum (x[i][j][t] for t in range(0, dim))) for i in range(0, dim) for j in range(0, dim) if i!=j), GRB.MINIMIZE)


##########################
#for i in range(1, dim):
#    for j in range(0, dim):
#        TSP_SP.addConstr(x[i][j][0] == 0)

#for i in range(1, dim):
#    for j in range(0, dim):
#        TSP_SP.addConstr(x[j][i][dim-1] == 0)
          
    
###########################

TSP_SP.addConstr(quicksum(x[0][j][0] for j in range(0, dim) if j!=0) == 1 )
           

for i in range(0, dim):
    if(i != 0):
        TSP_SP.addConstr((quicksum(x[i][j][1] for j in range(0, dim) if (i!=j and j!=0))) - x[0][i][0] == 0)
    

for i in range(0, dim):
    for t in range(2, dim-1):
        if(i != 0):
            TSP_SP.addConstr((quicksum(x[i][j][t] for j in range(0, dim) if (i!=j and j!=0))) - (quicksum(x[j][i][t-1] for j in range(0, dim) if (i!= j and j!=0))) == 0)

            
for i in range(0, dim):
    if(i != 0):
        TSP_SP.addConstr(x[i][0][dim-1] - (quicksum(x[j][i][dim-2] for j in range(0, dim) if (i!=j and j!=0)))  == 0)
        

TSP_SP.addConstr(quicksum(x[i][0][dim-1] for i in range(0, dim) if i != 0) == 1)
        
for i in range(0, dim):
    if(i != 0):
        TSP_SP.addConstr(quicksum(quicksum(x[i][j][t] for j in range(0, dim) if (i!=j and j!= 0)) for t in range(1, dim-1)) + x[i][0][dim-1]  <= 1)
        
TSP_SP.optimize()

executionTime = (time.time() - startTime)
print("The execution time is")
print(executionTime)

```
