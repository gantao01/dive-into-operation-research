```python
def Preprocess(file):
    array = []
    for line in f:
        array.append(line)
    xc = [0]*300
    yc = [0]*300
    maxrows = 300
    maxcols = 300
    Cost = [([0] * maxcols) for row in range(maxrows) ]

    dim = int(array[0])
    for it in range(1, dim+1):
        p = array[it]
        tokens = p.split()
        xc[it-1] = float(tokens[0])
        yc[it-1] = float(tokens[1])
    
    for i in range(0, dim):
        for j in range(0, dim):
            Cost[i][j] = pow((pow(xc[i]-xc[j], 2) + pow(yc[i] - yc[j], 2)), 0.5)
        
    for i in range(0, dim):
        for j in range(0, dim):
            if(i==j):
                Cost[i][j] = 99999
    return Cost, xc, yc, dim
```

```python

def findcomponents(x):
    component = 0
    for i in range(0, maxrows):
        if (visited[i] == 0):
            j = 0
            dfs(x, i, component, j)
            component = component +1

def dfs(M, i, l, j):
    visited[i] = 1
    components[l][j] = i
    #j = j+1
    for k in range(0, maxrows):
        if(M[i][k] == 1 and visited[k] != 1):
            dfs(M, k, l, j+1)
```

```python
from gurobipy import *
import string
import time

f = open("C:\\Users\\Prasangsha\\Downloads\\TSP_instance_n_30_s_1.dat")
Cost, xc, yc, dim = Preprocess(f)

visited = [0]*dim
maxrows = dim
maxcols = dim

components = [([-1] * maxcols) for row in range(maxrows) ]

TSP_Plain = Model('TSP_DFJ_PlainLoop')

x = [[0 for i in range(dim)] for j in range(dim)] 
for i in range(0, dim):
    for j in range(0, dim):
        x[i][j] = TSP_Plain.addVar(vtype=GRB.BINARY, name = "x[][]".format(i,j))


TSP_Plain.update()

TSP_Plain.setObjective(quicksum(quicksum(Cost[i][j]*x[i][j] for i in range(dim)) for j in range(dim)), GRB.MINIMIZE)


for i in range(0, dim):
    TSP_Plain.addConstr(quicksum(x[i][j] for j in range(0, dim) ) == 1)
    TSP_Plain.addConstr(quicksum(x[j][i] for j in range(0, dim) ) == 1)
    

startTime = time.time()
while(True):
    for i in range(0, dim):
        visited[i] = 0
    components = [([-1] * dim) for row in range(dim) ]
    TSP_Plain.optimize()
    Optim = x
    xoy = [[0 for i in range(dim)] for j in range(dim)]
    for i in range(dim):
        for j in range(dim):
            if(x[i][j].x == 1.0):
                xoy[i][j] = 1
            
    findcomponents(xoy)
    number_comp = 0
    for l in range(0, dim):
        if(components[l][0] != -1):
            number_comp = number_comp+1
    if(number_comp > 1):
        for l in range(0, number_comp):
            len_c = 0
            temp = [-1]*dim
            for j in range(0, dim):
                if (components[l][j] != -1):
                    temp[j] = components[l][j]
                    len_c = len_c +1
            TSP_Plain.addConstr(quicksum(x[temp[i]][temp[k]] for i in range(0, dim) for k in range(0, dim) if (i !=k and temp[i] != -1 and temp[k] != -1)) <= len_c - 1)
            TSP_Plain.update()
    else:
        break
executionTime = (time.time() - startTime)
print("The execution time is")
print(executionTime)
```
